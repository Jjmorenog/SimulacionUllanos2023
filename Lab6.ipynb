Tierra, 9.8m/s2 gravedad.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation, rc
from IPython.display import HTML

delta = 0.01
t_dur = 10
n_steps = int(t_dur/delta)
print(n_steps)
t=np.linspace(0,t_dur,n_steps)
a = np.zeros((n_steps, 2)) # Sin aceleración en X
a[:, 1] = -9.8 # Gravedad 9.8 m/s^2

# Solicitar al usuario que ingrese la posición inicial
s0 = np.array([float(input("Ingrese la posición inicial en x: ")), float(input("Ingrese la posición inicial en y: "))])

# Verificar que la posición inicial no esté fuera de las paredes establecidas
if s0[0] < 0 or s0[0] > 20:
    print("La posición inicial en x debe estar entre 0 y 20.")
    exit()
if s0[1] < 0 or s0[1] > 15:
    print("La posición inicial en y debe estar entre 0 y 15.")
    exit()

# Solicitar al usuario que ingrese la velocidad inicial
v0 = np.array([float(input("Ingrese la velocidad inicial en x: ")), float(input("Ingrese la velocidad inicial en y: "))])

delta_t = delta
all_v = np.zeros((n_steps, 2))
all_s = np.zeros((n_steps, 2))
v=v0
s=s0
umbral_velocidad = 0.001
damping_factor = 0.9  # Factor de amortiguamiento inicial
for x in range(n_steps):
    s = s + v*delta_t
    v = v + a[x]*delta_t
    #
    all_v[x] = v
    all_s[x] = s
    if s[1]<=0:
        s[1] = 0
        v[1]=-v[1]*damping_factor
        if abs(v[1]) < umbral_velocidad or abs(v[0]) < umbral_velocidad:  # Comprueba si la velocidad es menor que el umbral
            v[0] = 0  # Si es así, establece la velocidad a cero
            v[1] = 0  # Si es así, establece la velocidad a cero
        damping_factor *= 0.9
    elif s[1]>=15:
        s[1] = 15
        v[1]=-v[1]*damping_factor
        damping_factor *= 0.9  # Disminuir el factor de amortiguamiento
    if s[0]<=0:
        s[0] = 0
        v[0]=-v[0]*damping_factor
        damping_factor *= 0.9  # Disminuir el factor de amortiguamiento
    elif s[0]>=20:
        s[0] = 20
        v[0]=-v[0]*damping_factor
        damping_factor *= 0.9  # Disminuir el factor de amortiguamiento


plt.subplot(1,3,1)
plt.plot(t, a)
plt.ylabel("Aceleracion")
plt.xlabel("Tiempo")
#plt.ylim(0,0.5)
plt.subplot(1,3,2)
plt.plot(t, all_v)
plt.xlabel("Velocidad")
plt.xlabel("Tiempo")
plt.subplot(1,3,3)
plt.plot(t, all_s)
plt.ylabel("Posicion")
plt.xlabel("Tiempo")
plt.plot()
plt.suptitle('Dinámica del sistema')
plt.tight_layout()

# Ahora la animación:
fig, ax = plt.subplots()
ax.set_xlim(( 0, 25))
ax.set_ylim((0, 20))
line, = ax.plot([], [], lw=2)

# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    return (line,)

# animation function. This is called sequentially
def animate(i):
    x = [all_s[i][0]]
    y = [all_s[i][1]]
    line.set_marker('h')
    line.set_data(x, y)
    return (line,)

# call the animator. blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, init_func=init,
frames=n_steps, interval=10, blit=True)
HTML(anim.to_html5_video())
```
